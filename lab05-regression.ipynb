```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
df2 = pd.read_csv('Credit.csv')
df2


# sns.distplot(pd.DataFrame(df2['Age']).values)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# plt.scatter(df2['Income'], df2['Limit'])
# sns.scatterplot('Income', 'Limit', data = df2)

fig, axs = plt.subplots(ncols=2, figsize=(12,5))
sns.scatterplot(x='Income', y='Limit', data=df2, ax=axs[0])
sns.residplot(x='Income', y='Limit', data=df2, ax=axs[1])

sns.lmplot(x='Income', y='Limit', data=df2)

reg = LinearRegression()
reg.fit(pd.DataFrame(df2['Income']), pd.DataFrame(df2['Limit']))
print("slope: ", reg.coef_)
print("intercept: ", reg.intercept_)

# this data can apply SLR
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(12,5))
sns.scatterplot(x='Rating', y='Limit', data=df2, ax=axs[0])
sns.residplot(x='Rating', y='Limit', data=df2, ax=axs[1])

sns.lmplot(x='Rating', y='Limit', data=df2)
reg = LinearRegression()
reg.fit(pd.DataFrame(df2['Rating']), pd.DataFrame(df2['Limit']))
print("slope: ", reg.coef_)
print("intercept: ", reg.intercept_)


# this data can apply SLR
```
